```python
# import matplotlib.pyplot as plt


# # Data to plot
# labels = ['Wether Condition', 'Date/Time', 'Temp_C']
# sizes = [50, 30, 20]  # Proportion of each room type
# colors = ['#ff9999','#66b3ff','#99ff99']  # Colors for each segment
# explode = (0.1, 0, 0)  # explode 1st slice (Single Room)

# # Plotting the pie chart
# plt.figure(figsize=(8, 6))  # Set the figure size
# plt.pie(sizes, explode=explode, labels=labels, colors=colors,
#         autopct='%1.1f%%', shadow=True, startangle=140)

# plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# plt.title("Room Type Distribution")
# plt.show()


from fastapi import FastAPI
from fastapi.responses import StreamingResponse
import matplotlib.pyplot as plt
import io
import threading

app = FastAPI()

@app.get("/pie-chart")
async def get_pie_chart():
    labels = ['Weather Condition', 'Date/Time', 'Temp_C']
    sizes = [50, 30, 20]  # Proportion of each data type
    colors = ['#ff9999', '#66b3ff', '#99ff99']  
    explode = (0.1, 0, 0) 

    fig, ax = plt.subplots(figsize=(8, 6))
    ax.pie(sizes, explode=explode, labels=labels, colors=colors,
           autopct='%1.1f%%', shadow=True, startangle=140)
    ax.axis('equal')  
    plt.title("Data Type Distribution")

    buf = io.BytesIO()
    plt.savefig(buf, format='png')
    buf.seek(0)
    plt.close(fig)

    return StreamingResponse(buf, media_type="image/png")

def run_app():
    import uvicorn
    uvicorn.run(app, host="127.0.0.1", port=8000)
thrad=threading.Thread(target=run_app)
thrad.start()
```
